```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data['totalprice'],bins=20)
plt.title("distribution of total price")
plt.xlabel("total price")
plt.ylabel("number of orders")
plt.show()
```

Most orders have a total price below ₹1000. The distribution is right-skewed, indicating that high-value orders are less frequent.

```python
import seaborn as sns

sns.boxplot(x=data['totalprice'])
plt.title("Boxplot of Total Price")
plt.show()
```

A few high-value outliers are present, suggesting that majority of customers  placed low-cost orders, some customers placed significantly larger orders than the majority.

```python
plt.figure(figsize=(8,5))
product_sales=data.groupby('product')['totalprice'].sum()
product_sales.plot(kind='bar')
plt.title('product wise sales')
plt.show()
```

Printer generated the highest total sales, while Desk and Phone contributed comparatively lower sales.

```python
plt.figure(figsize=(8,5))
data['couponcode'].value_counts().plot(kind='bar')
plt.title('coupon code usage')
plt.show()
```

FREESHIP was the most frequently used coupon code, indicating that customers preferred free shipping offers.

```python
data['date'] = pd.to_datetime(data['date'])

monthly_sales = data.groupby(data['date'].dt.to_period('M'))['totalprice'].sum()

monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

Monthly sales fluctuated over time, with the highest sales occurring around October 2023 and the lowest around April 2023 and July 2024.

```python
plt.scatter(data['unitprice'],data['totalprice'])
plt.title('unitprice vs totalprice')
plt.show()
```

A strong positive relationship exists between Unit Price and Total Price. As unit price increases, total price also increases, with quantity creating distinct diagonal patterns.
