```python
import pandas as pd
import sqlite3

df = pd.read_excel("cleanedD.xlsx")

conn = sqlite3.connect("project.db")
df.to_sql("sales", conn, if_exists="replace", index=False)
```

```python
query = """
SELECT *
FROM sales;
"""

result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT *
FROM sales
LIMIT 5;
"""

result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT SUM([totalprice]) AS Total_Revenue
FROM sales;
"""

result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT AVG([unitprice]) AS Average_Unit_Price
FROM sales;
"""

result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT Product,
       SUM([totalprice]) AS Revenue
       FROM sales
       GROUP BY Product
       ORDER BY Revenue DESC;
       """
result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT [couponcode],
       COUNT(*) AS Times_Used
       FROM sales
       GROUP BY [couponcode];
       """

result = pd.read_sql(query, conn)
print(result)
```

```python
query = """
SELECT *
FROM sales
ORDER BY [totalprice] DESC
LIMIT 1;
"""

result = pd.read_sql(query, conn)
print(result)
```
